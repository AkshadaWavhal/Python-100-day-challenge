```python
a=[1,2,3,4,5]
l=a[0]
key=39
flag=False
h=len(a)-1
while l<=h:
    mid=(l+h)//2
    if(a[mid]==key):
        print("key present")
        flag=True
        break
    elif(a[mid]>key):
        h=mid-1
    else:
        l=mid+1
if not flag:
    print("key not present")
    
    
```

```python
# bubblle sort
a=[5,3,83,1,4]
n=len(a)
for i in range(n-1):
    for j in range(n-i-1):
        if(a[j]>a[j+1]):
            a[j],a[j+1]=a[j+1],a[j]
print(a)
```

```python
#selection sort
a=[7,5,7,2,3]
n=len(a)
for i in range(n-1):
    min_index=i
    for j in range(i+1,n):
        if(a[j]<a[min_index]):
            min_index=j
    a[min_index],a[i]=a[i],a[min_index]
print(a)
```
