```python
n=1234
sum=0
while n>0:
    r=n%10
    n=n//10
    sum+=r
print(sum)
```

```python
num = 7
is_prime = True

if num <= 1:
    is_prime = False
else:
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print(num, "is a Prime number")
else:
    print(num, "is NOT a Prime number")
```
