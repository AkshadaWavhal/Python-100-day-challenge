```python
# Sum of List Elements
a=[1,2,3,4]
sum=0
for i in a:
    sum+=i;
print(sum)
```

```python
# Factorial using Loop
fact=1
for i in range(1,6):
    fact*=i
print(fact)
```

```python
#  Check Palindrome
a="1223321"
a1=a[::-1]
print(a1)
if(a1==a):
    print("pallindrom")
else:
    print("not pallindrone")
```

```python
# prime number
num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
else:
    print("Not Prime")
```
