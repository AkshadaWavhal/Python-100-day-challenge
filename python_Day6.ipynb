```python
# 1.print pattern question 
for i in range(5):
    for j in range(5):
        print("* ",end=" ")
    print( "\n")
```

```python
# 2️⃣ Right-Angle Triangle
for i in range(1,5):
    print("*"*i)
    i+=1
    
```

```python
#3️⃣ Inverted Right-Angle Triangle
for i in range(5,0,-1):
    print("*" * i)
    
    
```

```python
#4 print number like serial
for i in range(1,5):
    for j in range(1,i+1):
        if(i>=j):
            print(j,end="")
    print( "\n")

```

```python
# 5 Repeating Row Numbers
for i in range(5):
    for j in range(1,i+1):
        if(i>=j):
            print(i,end="")
    print("\n")
```
