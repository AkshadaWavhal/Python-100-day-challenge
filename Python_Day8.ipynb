```python
n = 4
num = 1
for i in range(1, n+1):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```

```python
n = 4
for i in range(n):
    print(" "*(n-i-1) + "*"*(2*i+1))
for i in range(n-2, -1, -1):
    print(" "*(n-i-1) + "*"*(2*i+1))
```

```python
n = 5
for i in range(n):
    for j in range(n):
        if i==0 or i==n-1 or j==0 or j==n-1:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

```python
n = 4
for i in range(n):
    for j in range(i+1):
        print(chr(65 + j), end=" ")
    print()
```

```python
n = 4
for i in range(n):
    # Spaces
    print(" " * (n-i-1), end="")
    
    # Increasing alphabets
    for j in range(i+1):
        print(chr(65+j), end="")
        
    # Decreasing alphabets
    for j in range(i-1, -1, -1):
        print(chr(65+j), end="")
    print()
```
